This notebook analyzes spending trends using five datasets:
- **Yearly Spending Trend**
- **Average Spending by Gender**
- **Monthly Spending Trend**
- **Total Transactions by Category**
- **Total Transactions by Month**

We will explore, visualize, and summarize insights from these datasets.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

# Load datasets
yearly_spending = pd.read_csv('Yearly Spending Trend.csv')
gender_spending = pd.read_csv('Average Spending by Gender.csv')
monthly_spending = pd.read_csv('Monthly Spending Trend.csv')
top_10_cust = pd.read_csv('Top 10 Customers by Total Spent.csv')
top_10_merch = pd.read_csv('Top 10 Merchants by Total Sales.csv')
trans_cat = pd.read_csv('Total Transactions by Category.csv')
trans_month = pd.read_csv('Total Trasnasctions by Month.csv')
```

```python
# Preview datasets
display(yearly_spending.head())
display(gender_spending.head())
display(monthly_spending.head())
display(top_10_cust.head())
display(top_10_merch.head())
display(trans_cat.head())
display(trans_month.head())
```

## Yearly Spending Trend

```python
plt.figure(figsize=(8,5))
sns.lineplot(data=yearly_spending, x='year', y='monthly_spent', marker='o')
plt.title('Yearly Spending Trend')
plt.ylabel('Total Spending')
plt.xlabel('Year')
plt.grid(True)
plt.show()
```

## Monthly Spending Trend

```python
plt.figure(figsize=(10,6))
sns.lineplot(data=monthly_spending, x='month_name', y='monthly_spent', marker='o')
plt.title('Monthly Spending Trend')
plt.ylabel('Spending')
plt.xlabel('Month')
plt.xticks(rotation=45)
plt.show()
```

## Average Spending by Gender

```python
plt.figure(figsize=(7,5))
sns.barplot(data=gender_spending, x='Gender', y='avg_spending', palette='Set2')
plt.title('Average Spending by Gender')
plt.ylabel('Average Spending')
plt.xlabel('Gender')
plt.show()
```

## Top 10 Customers by Total Spent

```python
sns.barplot(data=top_10_cust, x='Name', y='total_spent', palette='Set3')
plt.title('Top 10 Customers by Total Spending')
plt.ylabel('Total Spent')
plt.xlabel('Name')
plt.xticks(rotation = 45)
plt.show()
```

## Top 10 Merchants by Total Sales

```python
sns.barplot(data=top_10_merch, x='Merchant Name', y='total_sales', palette='Set1')
plt.title('Top 10 Merchants by Total Sales')
plt.ylabel('Total Sales')
plt.xlabel('Merchant Name')
plt.xticks(rotation = 45)
plt.show()
```

## Transactions by Category

```python
plt.bar(trans_cat["Category"], trans_cat["total_transactions"], color="skyblue")
plt.xticks(rotation=45, ha="right")
plt.title("Total Transactions by Category")
plt.ylabel("Number of Transactions")
plt.xlabel("Category")
plt.tight_layout()
plt.show()
```

## Transactions by Month

```python
months_order = ["January","February","March","April","May","June",
                "July","August","September","October","November","December"]
month_df_sorted = trans_month.set_index("month_name").reindex(months_order).reset_index()
plt.plot(month_df_sorted["month_name"], month_df_sorted["total_transactions"], marker="o", color="green")
plt.title("Total Transactions by Month")
plt.ylabel("Number of Transactions")
plt.xlabel("Month")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Insights & Summary
- The **yearly trend** shows overall total spending per year.
- The **monthly breakdown** highlights which months have the highest/lowest spending.
- The **gender analysis** compares average spending between male and female customers.
- The **transactions by category** reveal which types of purchases are most frequent and where the highest spending occurs.
- The **transactions by month** show seasonal patterns, highlighting peaks and troughs in customer activity.
